8. Напишите функцию `validate()`, которая проверяет соответствие всех строк из листа `Отзывы` следующим правилам:
    * Рейтинг - это число от 0 до 5 включительно
    * Соответствующий рецепт имеется на листе `Рецепты`
    
В случае несоответствия этим правилам, выделите строку красным цветом

```python
def validate(sheet, reviews_sample, recipes_sample_with_tags_ingredients_new):
    for i in range(2, (reviews_sample_new.shape[0])+2):
        if (int(sheet.range((i, 5)).value) >=0) and (int(sheet.range((i, 5)).value <=5)) and (int(sheet.range((i,3)).value in recipes_sample_with_tags_ingredients_new)):                                                                                
            pass
        else:
            for j in range(1,9):
                sheet.range((i,j)).color = (220, 20, 60)

validate(sheet, reviews_sample, recipes_sample_with_tags_ingredients_new)
```
